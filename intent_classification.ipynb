```python
from sentence_transformers import SentenceTransformer

model_name = 'all-MiniLM-L6-v2'
sentence_transformer_model = SentenceTransformer(model_name)
```

```python
data = pd.read_csv("intent.csv")
texts = data["texts"].tolist()
```

```python
# Obtaining sentence embeddings using the Sentence Transformer model
dataset_embeddings = sentence_transformer_model.encode(texts)
```

```python
df_embeddings = pd.DataFrame(dataset_embeddings)
df_embeddings.to_csv('intent_embeddings.csv', index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from numpy import np

def classify_text(new_text):
  # Get new text embedding
  new_text_embedding = sentence_transformer_model.encode(new_text)

  # Calculate cosine similarity between new text and all dataset texts
  cosine_scores = cosine_similarity(new_text_embedding.reshape(1, -1), dataset_embeddings)

  # Find the most similar text in the dataset
  most_similar_idx = np.argmax(cosine_scores)
  print(most_similar_idx)

  # Retrieve the corresponding intent label from the dataset
  predicted_intent = data["intents"].iloc[most_similar_idx]
  return predicted_intent
```

```python
# Example usage
new_text = "how can i manage my money in share"
predicted_intent = classify_text(new_text)
print(f"Predicted intent for '{new_text}': {predicted_intent} ")
```
